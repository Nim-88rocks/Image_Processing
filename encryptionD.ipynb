## NORMALIZING THE IMAGE

```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

image=cv2.imread("lenac.jpg")
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 

# converting the image data type to float
converted_image=gray_image.astype(float)
converted_image

# calculating the minimum value
min_value=np.min(converted_image)

#  calculating the maximum value
max_value=np.max(converted_image)


# NORMALIZING THE IMAGE
normalized_image=((converted_image- min_value) / (max_value - min_value))*16


# displaying the original image
cv2.imshow("original_image",gray_image)
cv2.waitKey(0)
cv2.destroyAllWindows()

# displayoing the converted image
cv2.imshow("normalized_image",normalized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()

#dislpaying the images together
plt.figure(figsize=(10, 5))
original_image_saved=mpimg.imread('original_image_saved.png')
normalized_image_saved=mpimg.imread('normalized_image_saved.png')
plt.subplot(1, 2, 1)
plt.imshow(original_image_saved)
plt.axis('off')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(normalized_image_saved)
plt.axis('off')
plt.title('normalized_image')
```

## ENCRYPTING THE IMAGE

```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt

image=cv2.imread("lenac.jpg")
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 


# GENERATING A KEY MATRIX
random_matrix=np.random.rand(512,512)

#converting to float

random_matrix=random_matrix*(256)
random_matrix=random_matrix.astype(int)

# Perform element-wise addition of the input image and the key matrix
Encrypted_image=(gray_image + random_matrix) % 256
plt.imshow(Encrypted_image,cmap="gray")
```

## DECRYPTING THE IMAGE

```python
decrypted_image=(Encrypted_image-random_matrix)%256
plt.imshow(decrypted_image,cmap="gray")
```

```python
# CHECKIONG FOR CORRECTNESS
Correctness=[gray_image==decrypted_image]
tcount=0
fcount=0
for i in np.nditer(Correctness):
    if(i):
        tcount=tcount+1
    else:
        fcount=fcount+1
print("tcount = ",tcount,"fcount = ",fcount)    

""" IF THE NUMBER OF TRUES IS EQUAL TO THE SIZE OF THE MATRIX IT MEANS WE HAVE 0 ERROR """
""" HERE 512 x 512 = 262144 """
```

## DISPLAYING THE ORIGINAL, ENCRYPTED AND DECRYPTED IMAGES

```python
fig, axs = plt.subplots(1, 3, figsize=(12, 4))

# Display the images on the subplots
axs[0].imshow(gray_image, cmap='gray')
axs[0].axis('off')
axs[0].set_title('original_image')

axs[1].imshow(Encrypted_image, cmap='gray')
axs[1].axis('off')
axs[1].set_title('encrypted_image')

axs[2].imshow(decrypted_image, cmap='gray')
axs[2].axis('off')
axs[2].set_title('decrypted_image')
```
